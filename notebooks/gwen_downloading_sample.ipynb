```python
import kagglehub
from kagglehub import KaggleDatasetAdapter, load_dataset
import pandas as pd
import os
import requests
```

# Load the dataset and clean it

```python
df_sample = pd.read_csv("data_sampling200_topstyles.csv")
```

```python
df_sample.head()
```

```python
list_files = df_sample['img'].tolist()
list_files
```

```python
# Dossier de destination
destination_folder = "../raw_data/sample"
os.makedirs(destination_folder, exist_ok=True)

# Liste des URLs à télécharger
image_urls = list_files

for url in image_urls:
    file_name = url.split("/")[-1]  # récupère le nom de fichier depuis l'URL
    file_path = os.path.join(destination_folder, file_name)
    
    try:
        response = requests.get(url, timeout=10)  # timeout pour éviter de bloquer
        response.raise_for_status()  # déclenche une exception si le statut n'est pas 200
        with open(file_path, "wb") as f:
            f.write(response.content)
        print(f"Downloaded : {file_name}")
    except requests.exceptions.RequestException as e:
        print(f"Impossible to download : {file_name}. Reason : {e}")

print("Images are downloaded.")
```
